```python
import matplotlib.pyplot as plt
import matplotx
import numpy as np
import pandas as pd
```

```python
xls = pd.ExcelFile("../data/L8_OLI_RSR.xlsx")
bands = {}
band_names = [
    "CoastalAerosol",
    "Blue",
    "Green",
    "Red",
    "NIR",
    "Cirrus",
    "SWIR1",
    "SWIR2",
    "Pan",
]
for i, sheet in enumerate(band_names):
    df = pd.read_excel(xls, sheet)
    bands[sheet] = df.to_numpy()[:, :2]
```

```python
plt.figure(figsize=(12, 4))
with plt.style.context(matplotx.styles.dufte):
    for band, val in bands.items():
        plt.plot(val[:, 0], val[:, 1], label=band)

    plt.xlabel("Wavelength [nm]")
    plt.ylabel("Relative spectral response")
```
